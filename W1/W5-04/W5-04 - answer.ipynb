```python
# Use Selenium 
from selenium import webdriver
import chromedriver_binary  # Adds chromedriver binary to path
from webdriver_manager.chrome import ChromeDriverManager
import time
from bs4 import BeautifulSoup
url='https://public.flourish.studio/visualisation/1641110/embed?auto=1'


driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
data=[]
# page_number = driver.find_element_by_xpath('/html/body/main/section[4]/div[1]/div/div[4]/p/span').text
i = 1

while i < 32:
    # load data into bs4
    soup = BeautifulSoup(driver.page_source,'html.parser')

    # get data simply by looking for each a links
    for tr in soup.find_all('div',attrs={'class':'tr body-row'}):
        tokens = tr.text.split('\n')
        data.append(tokens)

    print("page: ", i)
    driver.find_element_by_xpath('/html/body/main/section[4]/div[1]/div/div[4]/button[2]').click()
    i += 1


data
```

```python
import re
data_within_range = []
for i in data:
    # check year first
    if int(re.search(r'-(\d+)', i[0])[1]) == 21:
        if re.search(r'-(\w+)-', i[0])[1] == 'Jan' or re.search(r'-(\w+)-', i[0])[1] == 'Feb' :
            data_within_range.append(i)
        elif re.search(r'-(\w+)-', i[0])[1] == 'Mar':
            if int(re.search(r'(\d+)-', i[0])[1]) < 21:
                data_within_range.append(i)
```
